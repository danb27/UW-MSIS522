```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

# include the plots in the notebook
%matplotlib inline

# set random seed to reproduce the result
np.random.seed(0)
```

## Line plot

```python
a = np.linspace(0, 10, 100)
b = np.exp(-a)

# matplotlib version
plt.plot(a, b)
plt.show()

# seaborn version
sns.lineplot(a, b)
```

## Scatter plot

```python
data = {'a': np.arange(50),
        'b': np.random.randn(50)*10,
        'color': np.random.randint(0, 50, 50),
        'size': np.abs(np.random.randn(50)*100)}

# matplotlib version
plt.scatter('a', 'b', c='color', s='size', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()

# seaborn version
sns.scatterplot('a', 'b', hue='color', size='size', data=data, legend='brief')
```

## Load the iris dataset from seaborn

```python
iris = sns.load_dataset("iris")
display(iris.head())
```

## Scatter plot with bivariate distribution

```python
sns.jointplot(x='sepal_length', y='sepal_width', data=iris)
```

## Histogram

```python
fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.distplot(iris.sepal_length, bins=10)
plt.subplot(1, 3, 2)
sns.distplot(iris.sepal_length, bins=20)
plt.subplot(1, 3, 3)
sns.distplot(iris.sepal_length, bins=10, kde=False)
plt.show()
```

## Boxplot

```python
sns.catplot(x="species", y="petal_length", kind="box", data=iris)
```

## Pair plot
To plot multiple pairwise bivariate distributions in a dataset, you can use the pairplot() function. This creates a matrix of axes and shows the relationship for each pair of columns in a DataFrame. by default, it also draws the univariate distribution of each variable on the diagonal Axes:

```python
sns.pairplot(iris, hue='species', height=2.5);
```
